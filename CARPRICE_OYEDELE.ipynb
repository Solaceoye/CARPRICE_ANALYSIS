#### Defining terms 

#### Compressionratio: The compression ratio is the ratio between the maximum and minimum volume during the compression stage of the power cycle in a piston	

#### Horsepower:Horsepower (hp) is a unit of measurement of power, or the rate at which work is done, usually in reference to the output of engines or motors.	

#### Peakrpm:RPM stands for revolutions per minute, and it's used as a measure of how fast any machine is operating at a given time.

#### City MPG: City driving in Carlsbad can be tough on your car because you have to stop frequently at stop signs and red lights. This means your vehicle’s speeds are inconsistent, which leads to lower fuel economy.

#### Highwaympg Highway driving is a lot easier on your vehicle because you generally stick to a consistent speed without a ton of stops along the way. That’s why highway MPG is almost always higher than city MPG.

CONCLUSION

The analysis of the provided dataset reveals several key relationships between various car attributes
and performance metrics. A strong correlation exists between engine size, horsepower, and curb weight,
indicating that more powerful engines tend to be heavier. 
Additionally, a positive relationship is observed between highway MPG and city MPG, 
suggesting that cars with better fuel efficiency in the city also perform well on the highway. 
Furthermore, the data suggests that cars with higher prices tend to have more advanced engine features,
such as fuel injection systems and higher compression ratios. 
The number of cylinders and engine size also appear to be related to the car's price. 
                                           
KEY TAKEAWAYS

1. Engine size, horsepower, and curb weight are strongly correlated. 
2. Highway MPG and city MPG are positively correlated. 
3. Car price is related to engine features, number of cylinders, and engine size. 
4. Drive wheel and engine location are related to car performance. 
These findings can inform car design and development, as well as consumer decision-making, 
by highlighting the complex interrelationships between various car attributes and performance metrics.
